```python
#Opens my webcam to start capturing a frame
cap = cv2.VideoCapture(0)

while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Our operations on the frame come here
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Display the resulting frame
    cv2.imshow('frame',gray)
    plt.imshow(gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
#loading video to frames
loadvideo = cv2.VideoCapture("Tyga - Switch Lanes  Feat The Game (Official Music.mp4")
while(loadvideo.isOpened()):
    ret , frame = loadvideo.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
loadvideo.release()
cv2.destroyAllWindows()
```

```python
# saving a video after capturing it on camera
video_to_save = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*"XVID")
out =cv2.VideoWriter("output2.mp4",fourcc , 20.0 ,(640,480))
while(video_to_save.isOpened()):
    ret, frame = video_to_save.read()
    if(ret==True):
        #frame = cv2.flip(frame,0)
        out.write(frame)
        cv2.imshow('frame', frame)
        if(cv2.waitKey(1) & 0xFF == ord('q')):
            break
    else:
        break

video_to_save.release()
out.release()
cv2.destroyAllWindows()
        
        
```
